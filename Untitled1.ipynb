```python
!pip install pandas
import pandas as pd
from google.colab import files
import numpy as np

# Function to simulate market prices
def simulate_market_price(base_price):
    variation = np.random.uniform(-0.2, 0.2)
    return round(base_price * (1 + variation), 2)

# Upload CSV file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Read the CSV file
products_df = pd.read_csv(file_name)
products_df.head()

# Add a simulated market price column
products_df['market_price'] = products_df['our_price'].apply(simulate_market_price)

def generate_insight(row):
    if row['our_price'] < row['market_price'] * 0.9:
        adjustment = round((row['market_price'] / row['our_price'] - 1) * 100, 2)
        return f"Consider raising price by ~{adjustment}%"
    elif row['our_price'] > row['market_price'] * 1.1:
        adjustment = round((1 - row['market_price'] / row['our_price']) * 100, 2)
        return f"Consider lowering price by ~{adjustment}%"
    else:
        return "Price is competitive"

products_df['recommendation'] = products_df.apply(generate_insight, axis=1)

increase_price = products_df[products_df['recommendation'].str.contains('raising')]
decrease_price = products_df[products_df['recommendation'].str.contains('lowering')]
competitive_price = products_df[products_df['recommendation'] == "Price is competitive"]

num_products = len(products_df)
num_increase = len(increase_price)
num_decrease = len(decrease_price)
num_competitive = len(competitive_price)

summary = (
    f"# Pricing Analysis Summary\n\n"
    f"- Total Products Analyzed: {num_products}\n"
    f"- Products Needing Price Increase: {num_increase} ({(num_increase / num_products) * 100:.2f}%)\n"
    f"- Products Needing Price Decrease: {num_decrease} ({(num_decrease / num_products) * 100:.2f}%)\n"
    f"- Products with Competitive Prices: {num_competitive} ({(num_competitive / num_products) * 100:.2f}%)\n\n"
)

report = summary + "# Detailed Recommendations\n\n"

for _, row in products_df.iterrows():
    report += (
        f"- **{row['product_name']}**:\n"
        f"  - Our Price: ${row['our_price']}\n"
        f"  - Market Price: ${row['market_price']}\n"
        f"  - Recommendation: {row['recommendation']}\n\n"
    )

# Report
with open("report.md", "w") as file:
    file.write(report)

files.download("report.md")

products_df.to_csv("updated_products.csv", index=False)

files.download("updated_products.csv")

print(summary)
print("Analysis complete. Detailed recommendations are saved in the report.md file.")
```
